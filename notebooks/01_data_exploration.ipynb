```python
import pandas as pd

# Load CSV files
users = pd.read_csv("../data/raw/AudioShots_UAT.users.csv")
likes = pd.read_csv("../data/raw/AudioShots_UAT.likes.csv")
listens = pd.read_csv("../data/raw/AudioShots_UAT.listeningprogresses.csv")
categories = pd.read_csv("../data/raw/AudioShots_UAT.categories.csv")
languages = pd.read_csv("../data/raw/AudioShots_UAT.languages.csv")

# Print shapes
print("Users:", users.shape)
print("Likes:", likes.shape)
print("Listening:", listens.shape)
print("Categories:", categories.shape)
print("Languages:", languages.shape)

print("\n")

print("USERS COLUMNS:")
print(users.columns.tolist())

print("\nLIKES COLUMNS:")
print(likes.columns.tolist())

print("\nLISTENING COLUMNS:")
print(listens.columns.tolist())

print("\nCATEGORIES COLUMNS:")
print(categories.columns.tolist())

print("\nLANGUAGES COLUMNS:")
print(languages.columns.tolist())

```
